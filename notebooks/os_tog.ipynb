# OS-TOG Framework

This notebook provides a demonstration on inferencing the OS-TOG framework using the database used for real-world experiments.


```python
# import standard libraries
import os
import sys
import torch
import cv2

# add root code directory to python path
current_dir = os.path.dirname(os.path.abspath(''))
print(f"[INFO] Adding '{current_dir}' to python path.")
sys.path.append(current_dir)

# import libraries from os_tog
from os_tog.framework import OS_TOG
from os_tog.config import OSTOGConfig as cfg

cfg.DEVICE = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')

framework = OS_TOG(cfg)
```

```python
target_object = "hammer"
target_task = "hammering"

scene_img = cv2.imread("../samples/example_scene.png")
scene_img = cv2.cvtColor(scene_img, cv2.COLOR_BGR2RGB)

grasps = framework.get_prediction(scene_img, target_object, target_task)
```
